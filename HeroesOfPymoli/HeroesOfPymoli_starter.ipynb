### I have been assigned the task of analyzing the Company data for their most recent fantasy game Heroes of Pymoli. Below is all the work associated with this task

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# HeroesOfPymoli file to Load
Game_Data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
Games_df = pd.read_csv(Game_Data)

# Inspecting contents for row and column values
Games_df.head() 
```

## Player Count

* Display the total number of players


```python
# Getting a total for number of players based on number of Players ID
total_number_of_players = len(Games_df["SN"].value_counts())
total_players_count = pd.DataFrame({"Total Players":[total_number_of_players]})
total_players_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the number of unique Items in the Dataframe
unique_items_count = len(Games_df["Item Name"].unique())

# Calculate the Average Price
Average_Price_count_float = Games_df["Price"].mean()
Average_Price_count = Average_Price_count_float
              
# Calculate the Number of Purchases
total_purchase_count = len(Games_df["Item Name"])

# Calculate the Total Revenue
total_revenue_count = Games_df["Price"].sum()

# Place all of the data found into a summary DataFrame\n",
summary_table = pd.DataFrame({"Number of Unique Items":[unique_items_count],
                             "Average Price":Average_Price_count,
                             "Number of Purchases":[total_purchase_count],
                             "Total Revenue":total_revenue_count})
summary_table["Average Price"] = summary_table["Average Price"].astype(float).map("${:,.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# determine the total number of players
Total_number_of_Players_count = len(Games_df["SN"].unique())

# Get list of Gender entries
gender_column = Games_df.groupby("Gender")

# Count the total of screen names "SN" by gender
Gender_counts = gender_column.nunique()["SN"]

# Total count by gender and divivde by total players 
Gender_percentage = round(Gender_counts / Total_number_of_Players_count * 100, 2)

# Format output 
Demographics_df = pd.DataFrame({"Total Count": Gender_counts, 
                                      "Percentage of Players": Gender_percentage})

#Format the data frame to remove index (ie Gender)
Demographics_df.index.name = None

Demographics_df.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Setup Bins and Labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150] # safe vet humans do not live to 150
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Add new column to dataframe 
Games_df["Age Group"] = pd.cut(Games_df["Age"],bins, labels=group_labels)
Games_df

# Link new column to new Dataframe
New_AGE_Group = Games_df.groupby("Age Group")

# Count total of non duplicated players 
total_count_age = New_AGE_Group["SN"].nunique()

# Calculate Percentage of Players 
percentage_by_age = (total_count_age/total_number_of_players) * 100

# Create data frame with obtained values
Purchase_age = pd.DataFrame({"Total Count": total_count_age, "Percentage of Players": percentage_by_age})

# Format the data frame with no index 
Purchase_age.index.name = None

# Format percentage  
Purchase_age.style.format({"Percentage of Players":"{:,.2f}%"})
```
